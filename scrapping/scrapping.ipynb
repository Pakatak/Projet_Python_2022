```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from datetime import *
import calendar
```

```python
def recup_data(query,limit=10000):

  tweets=[]

  for tweet in sntwitter.TwitterSearchScraper(query).get_items():
      if len(tweets)==limit:
          break
      else:
          tweets.append([tweet.date,tweet.user.username,tweet.content,tweet.retweetCount,tweet.likeCount])
  return pd.DataFrame(tweets,columns=['Datetime','User','Tweet','retweet','like'])
```

```python
def recup_all_data():
    df=pd.DataFrame(columns=['Datetime','User','Tweet','retweet','like'])
    tweets_csv_list = []
    for i in range(1,12):
      for j in range(1,28,13):
        query="(to:EmmanuelMacron) lang:fr until:2022-{}-{} since:2022-{}-{}".format(i,j+6,i,j)
        tweets_csv_list.append(recup_data(query))
    df = pd.concat(tweets_csv_list ,axis=0, sort=True).sort_values("Datetime", ignore_index=True)
    df['Month']=df['Datetime'].apply(lambda x: calendar.month_abbr[x.month])
    df['Date']=df['Datetime'].apply(datetime.date)
    df.to_csv("df.csv",index=False)
    return df
```
